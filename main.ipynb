```python
import pandas as pd
```

```python
data_url = "./adult.data.csv"
```

```python
data = pd.read_csv(data_url)
data.head()
```

```python
#Сколько женщин в данном наборе данных? Ответ: 10771
df = pd.DataFrame(data, columns=["sex"])
female_count = df[df["sex"] == "Female"].count()
male_count = df[df["sex"] == "Male"].count()
print(female_count)
print(male_count)
print(df.count())
```

```python
#Какой средний возраст женщин в данном наборе данных?   Ответ: 36.85 лет
female_df = pd.DataFrame(data, columns=["sex", "age"])
female_df = pd.DataFrame(female_df[female_df["sex"] == "Female"], columns=["age"])
female_df.mean()
```

```python
#Какова доля граждан Германии? Ответ: 0.42 %
df_all = pd.DataFrame(data, columns=["native-country"])
df_germany_count = df_all[df_all["native-country"] == "Germany"].count()
df_all_count = df_all.count()
print((df_germany_count / df_all_count) * 100)
```

```python
#**4-5. Каковы средние значения и среднеквадратичные отклонения возраста 
# тех, кто получает более 50K в год (признак salary) и тех, кто получает менее 50K в год? ** Средние значения <=50K 44.2 ; >50K 36.8
# Среднеквадратичное отклонение <=50K 10.5 ; >50K 14

df_salary_all = pd.DataFrame(data, columns=["age", "salary"])

df_salary_more = df_salary_all[df_salary_all["salary"] == "<=50K"]

df_salary_less = df_salary_all[df_salary_all["salary"] == ">50K"]

print("Среднее значение: " ,df_salary_less.loc[:, "age"].mean(), df_salary_more.loc[:, "age"].mean())

print("Среднеквадратичное отклонение: ", df_salary_less["age"].std(), df_salary_more["age"].std())
```

```python
# 6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? 
# (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate) 
# Ложь

df_education = pd.DataFrame(data, columns=["education", "salary"])
df_e_more50 = df_education[df_education["salary"] == ">50K"]

educations = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]

df_e_more50 = df_education[(~df_education["education"].isin(educations)) & (df_education["salary"] == ">50K")]

print(df_e_more50.count())

```

```python
# 7. Выведите статистику возраста для каждой расы (признак race) и каждого пола. 
# Используйте groupby и describe. Найдите таким образом максимальный возраст мужчин расы Amer-Indian-Eskimo.
# Максимальный возраст 82 года

df = pd.DataFrame(data, columns=["age", "race", "sex"])

df.groupby(["race", "sex"]).describe()
```

```python
#8. Среди кого больше доля зарабатывающих много (>50K): 
# среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, 
# у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), 
# остальных считаем холостыми. Ответ: Доля зарабатывающих больше среди женатых людей

df = pd.DataFrame(data, columns=["salary", "marital-status", "sex"])
df = df[df["sex"] == "Male"]

married = df[(df["marital-status"].str.startswith("Married"))]
married_salary_more = married[married["salary"] == ">50K"]
married_salary_less = married[married["salary"] == "<=50K"]
print("Доля зарабатывающих >50k женатых людей: ", (married_salary_more["salary"].count() / married["salary"].count()) * 100)

not_married = df[~df["marital-status"].str.startswith("Married")]
nmarried_salary_more = not_married[not_married["salary"] == ">50K"]
print("Доля зарабатывающих >50K не женатых: ", (nmarried_salary_more["salary"].count() / not_married["salary"].count()) * 100)
```

```python
#9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? 
# Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?
# 29% - доля зарабатывающих много

df = pd.DataFrame(data, columns=["hours-per-week", "salary"])

max_hpw = int(df["hours-per-week"].max())

mhpw = df[df["hours-per-week"] == max_hpw]

print(f"Наибольшее количество часов в неделю {max_hpw} и работают столько" , mhpw["hours-per-week"].count(), "человек", end="\n\n\n")


mhpw_salary_more = mhpw[mhpw["salary"] == ">50K"]

print((mhpw_salary_more["salary"].count() / mhpw["salary"].count()) * 100)

```
